# Define Ink Usage per Pixel:
Printers use CMYK, every pixel in the pdf will consume a combination of CMYK ink values depending on the color of the pixel. A pixel with (255, 255, 255) in RGB  shouldnâ€™t consume any ink, but for each non-white pixel should have its ink usage distributed across the CMYK channels.

Initialize unit counters for each color channel

# Run Unit Analysis

```python
import fitz 
import numpy as np
from PIL import Image

unit_counts = {
    "Cyan": 0,
    "Magenta": 0,
    "Yellow": 0,
    "Black": 0
}

def rgb_to_cmyk_units(r, g, b):
    c = 1 - (r / 255)
    m = 1 - (g / 255)
    y = 1 - (b / 255)
    k = min(c, m, y)
    c = max(0, c - k)
    m = max(0, m - k)
    y = max(0, y - k)
    return c, m, y, k

def process_image(image):
    global unit_counts
    image = image.convert("RGB")  # Ensure RGB format
    pixels = image.load()  # Load pixels
    
    width, height = image.size
    print(f"Image size: {width}x{height}")

    # Access pixels the and compute RGB to CMYK values
    for y in range(height):
        for x in range(width):
            r, g, b = pixels[x, y]  # Access RGB values
            c, m, y, k = rgb_to_cmyk_units(r, g, b)
            unit_counts["Cyan"] += c
            unit_counts["Magenta"] += m
            unit_counts["Yellow"] += y
            unit_counts["Black"] += k

def analyze_pdf(file_path):
    doc = fitz.open(file_path)
    for page_num in range(len(doc)):
        print(f"Processing page {page_num + 1}...")
        page = doc.load_page(page_num)
        pix = page.get_pixmap()
        
        # Convert pixmaps to image
        image = Image.frombytes("RGB", (pix.width, pix.height), pix.samples)
        
        process_image(image)
    
    doc.close()
    
    for color, count in unit_counts.items():
        print(f"{color} units: {count:.2f}")

analyze_pdf("sample.pdf")
```
